# Regression

Putting independent variables as x and dependent variable as y

```python
df['PowerTrain'].replace(to_replace=['RWD','AWD','FWD'],value=[0, 2,1],inplace=True)
x=df[['AccelSec','Range_Km','TopSpeed_KmH','Efficiency_WhKm', 'RapidCharge','PowerTrain']]
y=df['INR']
```

```python
import statsmodels.api as sm
x= sm.add_constant(x)
results = sm.OLS(y,x)
```

Fitting the model and summarizing

```python
model=results.fit()
model.summary()
```

```python
from sklearn.preprocessing import StandardScaler,PowerTransformer
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_validate,train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn import metrics
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans, MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
from collections import Counter
from sklearn.metrics import r2_score,silhouette_score,confusion_matrix,accuracy_score
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer, InterclusterDistance
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=42)
```

```python
lr= LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

Finding out the R-squared value

```python
r2=(r2_score(y_test,pred))
print(r2*100)
```

Around 58% of the dependant variable has been explained by the independant variables

Clustering

```python
df.drop('PriceEuro', axis=1, inplace=True)
```

```python
df['FastCharge_KmH'][91]=0
```

```python
df['FastCharge_KmH'] = pd.to_numeric(df['FastCharge_KmH'])
```

```python
features = ['AccelSec','TopSpeed_KmH','Efficiency_WhKm','FastCharge_KmH', 'RapidCharge','Range_Km', 'Seats', 'INR','PowerTrain']
# Separating out the features
x = df.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=9)
t = pca.fit_transform(x)
data2 = pd.DataFrame(t, columns=['PC1', 'PC2','PC3','PC4','PC5','PC6', 'PC7', 'PC8','PC9'])
data2
```

```python
df_9=data2.iloc[:,:9]
df_9.head(3)
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_9.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='coolwarm')
plt.show()
```

```python
linked = linkage(data2, 'complete')
plt.figure(figsize=(13, 9))
dendrogram(linked, orientation='top')
plt.show()
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
#K-means clustering 

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(t)
df['cluster_num'] = kmeans.labels_ 
print (kmeans.labels_) 
print (kmeans.inertia_)  
print(kmeans.n_iter_) 
print(kmeans.cluster_centers_) 
```

```python
Counter(kmeans.labels_)
```

```python
#Visulazing clusters
sns.scatterplot(data=data2, x="PC1", y="PC9", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

```python
X=data2[['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9']]
y=df['INR']
```
